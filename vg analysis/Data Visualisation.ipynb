```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/vgsales_cleaned.csv')
```

```python
df.columns
```

```python
# return the sales per year
def sales_by_year():
    sales_by_year = {}
    for year in np.sort(df.Year.unique()):
        sales = 0
        for index in df.index:
            if(df['Year'][index] == year):
                sales += df['Global_Sales'][index]
        sales_by_year[year] = np.round(sales, 2)
    return sales_by_year
```

```python
# return the sales per genre
def sales_by_genre():
    sales_by_genre = {}
    for genre in df.Genre.unique():
        sales = 0
        for index in df.index:
            if(df['Genre'][index] == genre):
                sales += df['Global_Sales'][index]
        sales_by_genre[genre] = np.round(sales, 2)
    return sales_by_genre
```

```python
# return the sales per platform
def sales_by_platform():
    sales_by_platform = {}
    for platform in df.Platform.unique():
        sales = 0
        for index in df.index:
            if(df['Platform'][index] == platform):
                sales += df['Global_Sales'][index]
        sales_by_platform[platform] = np.round(sales, 2)
    return sales_by_platform
```

```python
# return the sales per publisher
def sales_by_publisher():
    sales_by_publisher = {}
    for publisher in df.Publisher.unique():
        sales = 0
        for index in df.index:
            if(df['Publisher'][index] == publisher):
                sales += df['Global_Sales'][index]
        sales_by_publisher[publisher] = np.round(sales, 2)
    return sales_by_publisher
```

```python
# return the popularity of genre per year
def pop_by_genre(genre):
    pop_by_genre = {}
    for year in np.sort(df.Year.unique()):
        sales = 0
        for index, data in df.iterrows():
            if(data['Year'] == year and data['Genre'].lower() == genre.lower()):
                sales += data['Global_Sales']
        pop_by_genre[year] = np.round(sales, 2)
    return pop_by_genre
```

```python
# return the popularity of genre in each platform
def pop_by_platform(genre):
    pop_by_platform = {}
    for platform in df.Platform.unique():
        sales = 0
        for index, data in df.iterrows():
            if(data['Platform'] == platform and data['Genre'].lower() == genre.lower()):
                sales += data['Global_Sales']
        pop_by_platform[platform] = np.round(sales, 2)
    return pop_by_platform
```

```python
# return the popularity of genre for each publisher
def pop_by_publisher(genre):
    pop_by_publisher = {}
    for publisher in df.Publisher.unique():
        sales = 0
        for index, data in df.iterrows():
            if(data['Publisher'] == publisher and data['Genre'].lower() == genre.lower()):
                sales += data['Global_Sales']
        pop_by_publisher[publisher] = np.round(sales, 2)
    return pop_by_publisher
```

```python
year = sales_by_year()
x = list(year.keys())
y = list(year.values())

plt.figure(figsize=(15,8))
plot = sns.barplot(x, y)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
genre = sales_by_genre()
x = list(genre.keys())
y = list(genre.values())

plt.figure(figsize=(15,8))
plot = sns.barplot(x, y)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
platform = sales_by_platform()
x = list(platform.keys())
y = list(platform.values())

plt.figure(figsize=(15,8))
plot = sns.barplot(x, y)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
publisher = sales_by_publisher()
x = list(publisher.keys())
y = list(publisher.values())

plt.figure(figsize=(100,8))
plot = sns.barplot(x, y)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```

```python
# yearwise sales report of genre: Action
action_pop = pop_by_genre('action')
x = list(action_pop.keys())
y = list(action_pop.values())

plt.figure(figsize=(15,8))
plot = sns.plot(x, y)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.show()
```
